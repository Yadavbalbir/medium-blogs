### Understanding the `plot()` Function in Matplotlib

#### Key Points:

1. **Purpose of `plot()`**  
   - The `plot()` function is used to draw points (markers) in a diagram.

2. **Default Behavior**  
   - By default, the `plot()` function connects the points with a line.

3. **Parameters**  
   - The function takes two main parameters:
     - **Parameter 1**: An array containing the points on the X-axis.
     - **Parameter 2**: An array containing the points on the Y-axis.




```python
# Example: Line passing through multiple points

xpoints = [1, 4, 6, 8]
ypoints = [3, 5, 1, 10]

plt.plot(xpoints, ypoints)
plt.show()
```

```python
# Example: Default values of x 0, 1, 2, 3

ypoints = [3, 5, 1, 10, 4, 15, 4,10]

plt.plot(ypoints)
plt.show()
```

### Markers
You can use the `marker` keyword argument in Matplotlib's `plot()` function to highlight each point with a specified marker.

```python
# Example plot with 'o' marker

xpoints = [1, 4, 6, 8]
ypoints = [3, 5, 1, 10]

plt.plot(xpoints, ypoints, marker='X')
plt.show()
```

```python
# Example plot with 'X' marker

xpoints = [1, 4, 6, 8]
ypoints = [3, 5, 1, 10]

plt.plot(xpoints, ypoints, marker='D')
plt.show()
```

### Marker Size
You can use the keyword argument **markersize** or the shorter version, **ms** to set the size of the markers:

```python
# Example plot with 'X' marker

xpoints = [1, 4, 6, 8]
ypoints = [3, 5, 1, 10]

plt.plot(xpoints, ypoints, marker='X', ms='10')
plt.show()
```

## Linestyle
**linestyle**, or shorter **ls**, to change the style of the plotted line <br>

| Style   | Representation |
|---------|----------------|
| solid   | '-'            |
| dotted  | ':'            |
| dashed  | '--'           |
| dashdot | '-.'           |

```python
# Example plot with 'X' marker and 'dotted' linestyle

xpoints = [1, 4, 6, 8]
ypoints = [3, 5, 1, 10]

plt.plot(xpoints, ypoints, marker='X', ms='10', ls='-.')
plt.show()
```

### Line Width

You can use the keyword argument **linewidth** or the shorter **lw** to change the width of the line.

The value is a **floating number**, in points:



```python
# Example plot with 'X' marker, 'solid' linestyle, color red, width 3

xpoints = [1, 4, 6, 8]
ypoints = [3, 5, 1, 10]

plt.plot(xpoints, ypoints, marker='X', linestyle='solid', color='r', linewidth="5")
plt.show()
```

### `bar()` Function in Matplotlib

The `bar()` function in Matplotlib is used to create a bar chart. It takes the following arguments:

1. **Categories**: The first argument represents the categories along the x-axis. These can be passed as an array or list.
2. **Values**: The second argument represents the height or value of each bar, corresponding to the categories. This is also passed as an array or list.


```python
# Horizontal bar chat
# Data for the bar chart
categories = ['Math', 'Science', 'History', 'English', 'Art']  # Subjects
scores = [85, 90, 78, 88, 92]  # Scores for each subject

# Creating the bar chart
plt.barh(categories, scores, color='skyblue')

# Adding title and labels
plt.title("Student Scores in Different Subjects")
plt.xlabel("scores")
plt.ylabel("Subjects")

# Show plot
plt.show()
```

### Histogram

A **histogram** is a graphical representation of the **frequency distribution** of a dataset. It shows the number of observations within each given interval, known as **bins** or **classes**.

### Characteristics of a Histogram:
- **X-axis**: Represents the intervals (or bins) of the data.
- **Y-axis**: Represents the frequency (count) of data points that fall within each bin.
- The height of each bar shows the number of observations in that particular interval.

Histograms are useful for understanding the distribution of data, including the shape, spread, and central tendency.



```python
# Sample data: ages of a group of people
ages = [20, 25, 27, 30, 35, 40, 42, 45, 48, 50, 55, 60, 62, 65, 70, 72, 75, 80]

# Creating the histogram
plt.hist(ages, bins=10, color='hotpink', edgecolor='green')

# Adding title and labels
plt.title("Age Distribution")
plt.xlabel("Age Groups")
plt.ylabel("Frequency")

# Show plot
plt.show()
```

Histogram Explanation

- **`ages`**: A list of ages of a group of people. This dataset will be used to generate the histogram.
  
- **`plt.hist()`**: Creates the histogram with the specified data and bins. This function divides the data into intervals (bins) and shows the frequency of occurrences for each interval.
  
- **`bins=5`**: Divides the data into 5 intervals (bins). The number of bins determines how the data is grouped. A higher number of bins gives more detail, while fewer bins provide a broader view of the data distribution.
  
- **`color`**: Sets the color of the bars in the histogram. This allows for customization of the bar appearance to enhance visual appeal or readability.
  
- **`edgecolor`**: Sets the color of the edges of the bars. It outlines the bars to make them more distinct or visually appealing.


## Pie Chart
A pie chart is a useful tool for displaying categorical data and visualizing the proportion or percentage of each category within a whole.

```python
y = [35, 25, 25, 15]

plt.pie(y)
plt.show()
```

```python
y = [35, 25, 25, 15]
mylabels = ["Apples", "Bananas", "Cherries", "Mango"]

plt.pie(y, labels = mylabels)
plt.show()
```

```python
# puts apple wedge 0.2 away
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0.3, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

```python
# define color
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
mycolors = ["orange", "hotpink", "hotpink", "hotpink"]


plt.pie(y, labels = mylabels, explode = myexplode, colors=mycolors)
plt.show()
```

```python
# legends
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
mycolors = ["orange", "hotpink", "red", "green"]
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]


plt.pie(y, explode = myexplode, colors=mycolors, labels=mylabels)
plt.legend()
plt.show()
```
